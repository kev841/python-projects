```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
from sklearn.cluster import KMeans
```

```python
data = pd.read_csv('/content/Countries-exercise.csv')
data
```

```python
plt.scatter(data['Longitude'],data['Latitude'])
plt.ylim(-90,90)
plt.xlim(-180,180)
plt.show
```

```python
x = data.iloc[:,1:3]
```

```python
clusters = KMeans(8)
```

```python
clusters.fit(x)
```

```python
identidfied_with_clusters = clusters.fit_predict(x)
identidfied_with_clusters
```

```python
data_with_clusters = data.copy()
data_with_clusters['Clusters'] = identidfied_with_clusters
data_with_clusters
```

```python
plt.scatter(data_with_clusters['Longitude'],data_with_clusters['Latitude'],c=data_with_clusters['Clusters'],cmap='rainbow')
plt.xlim(-180,180)
plt.ylim(-90,90)
plt.show()
```
