```python
from docx import Document
import matplotlib.pyplot as plt

# Create a Word document
doc = Document()
doc.add_heading("Employee Data Cleaning & Analysis Report", 0)

# --- Dataset Overview ---
doc.add_heading("1. Dataset Overview", level=1)
buf = []
df.info(buf=buf)
info_str = "\n".join(buf)
doc.add_paragraph(info_str)

# Describe
desc = df.describe(include="all").to_string()
doc.add_paragraph("Summary Statistics:\n" + desc)

# --- Average Salary by Department ---
doc.add_heading("2. Average Salary by Department", level=1)
avg_salary = df.groupby("Department")["Salary"].mean()
doc.add_paragraph(avg_salary.to_string())

# --- Employee Count by Department ---
doc.add_heading("3. Employee Count by Department", level=1)
dept_counts = df["Department"].value_counts()
doc.add_paragraph(dept_counts.to_string())

# --- Average Age by Department ---
doc.add_heading("4. Average Age by Department", level=1)
avg_age = df.groupby("Department")["Age"].mean()
doc.add_paragraph(avg_age.to_string())

# --- Employees by Join Year ---
doc.add_heading("5. Employees by Join Year", level=1)
df["Join_Year"] = df["Join_Date"].dt.year
join_trends = df["Join_Year"].value_counts().sort_index()
doc.add_paragraph(join_trends.to_string())

# --- Add Charts ---
doc.add_heading("6. Visualizations", level=1)

# Salary Distribution
plt.figure()
df["Salary"].plot(kind="hist", bins=5, edgecolor="black", title="Salary Distribution")
plt.xlabel("Salary")
plt.savefig("salary_dist.png")
plt.close()
doc.add_picture("salary_dist.png")

# Join Year Trends
plt.figure()
join_trends.plot(kind="bar", title="Employees Joined per Year")
plt.xlabel("Year")
plt.ylabel("Number of Employees")
plt.savefig("join_trends.png")
plt.close()
doc.add_picture("join_trends.png")

# Save document
doc.save("Employee_Data_Report.docx")
print(" Report generated: Employee_Data_Report.docx")
```

```python
from docx import Document
import matplotlib.pyplot as plt
import io

# Create a Word document
doc = Document()
doc.add_heading("Employee Data Cleaning & Analysis Report", 0)

# --- Dataset Overview ---
doc.add_heading("1. Dataset Overview", level=1)
buf = io.StringIO()
df.info(buf=buf)
info_str = buf.getvalue()
doc.add_paragraph(info_str)

# Describe
desc = df.describe(include="all").to_string()
doc.add_paragraph("Summary Statistics:\n" + desc)

# --- Average Salary by Department ---
doc.add_heading("2. Average Salary by Department", level=1)
avg_salary = df.groupby("Department")["Salary"].mean()
doc.add_paragraph(avg_salary.to_string())

# --- Employee Count by Department ---
doc.add_heading("3. Employee Count by Department", level=1)
dept_counts = df["Department"].value_counts()
doc.add_paragraph(dept_counts.to_string())

# --- Average Age by Department ---
doc.add_heading("4. Average Age by Department", level=1)
avg_age = df.groupby("Department")["Age"].mean()
doc.add_paragraph(avg_age.to_string())

# --- Employees by Join Year ---
doc.add_heading("5. Employees by Join Year", level=1)
df["Join_Year"] = df["Join_Date"].dt.year
join_trends = df["Join_Year"].value_counts().sort_index()
doc.add_paragraph(join_trends.to_string())

# --- Add Charts ---
doc.add_heading("6. Visualizations", level=1)

# Salary Distribution
plt.figure()
df["Salary"].plot(kind="hist", bins=5, edgecolor="black", title="Salary Distribution")
plt.xlabel("Salary")
plt.savefig("salary_dist.png")
plt.close()
doc.add_picture("salary_dist.png")

# Join Year Trends
plt.figure()
join_trends.plot(kind="bar", title="Employees Joined per Year")
plt.xlabel("Year")
plt.ylabel("Number of Employees")
plt.savefig("join_trends.png")
plt.close()
doc.add_picture("join_trends.png")

# Save document
doc.save("Employee_Data_Report.docx")
print(" Report generated: Employee_Data_Report.docx")
```
